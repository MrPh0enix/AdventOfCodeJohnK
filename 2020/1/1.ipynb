```python
with open('input.txt', 'r') as file:
    inDoc = file.read()
    
inDoc = inDoc.split('\n')
inDoc = [int(x) for x  in inDoc]
inDoc
```

```python
def calculate(inDoc):

    for idx, i in enumerate(inDoc):
        for idx2, j in enumerate(inDoc[idx+1:]):
            for k in inDoc[idx+idx2+1:]:
                if i + j + k == 2020:
                    return i*j*k
            
print(calculate(inDoc))
```
