```python
with open('input.txt', 'r') as file:
    Doc = file.read()

Lis = Doc.split('\n')
inptLis = Lis
```

```python
comps = set()
compsPairs = set()
for line in inptLis:
    splits = line.split('-')
    compsPairs.add(tuple(sorted(splits)))
    comps.add(splits[0])
    comps.add(splits[1])
  


lanParties = set()   
for comp in comps:
    for pair in compsPairs:
        if tuple(sorted([comp, pair[0]])) in compsPairs and tuple(sorted([comp, pair[1]])) in compsPairs:
            lanParties.add(tuple(sorted([comp, *pair])))
            

            
print(lanParties)

lanParties_witht =  set()
for i in lanParties:
    for comp in i:
        if 't' == comp[0]:
            lanParties_witht.add(tuple(sorted(i)))
            break

print(lanParties_witht)

print(len(lanParties_witht))
```

```python
input = []
for i in  inptLis:
    input.append(i.split('-'))

conns = {}
for x, y in input:
    if x not in conns.keys(): conns[x] = set()
    if y not in conns.keys(): conns[y] = set()
    conns[x].add(y)
    conns[y].add(x)
    

lanParties = set()   

def recur_search(comp, currSeq):
    ad = tuple(sorted(currSeq))
    if ad in lanParties:
        return
    lanParties.add(ad)
    for next in conns[comp]:
        if next in currSeq:
            continue
        if not all(next in conns[node] for node in currSeq): 
            continue
        recur_search(next, {*currSeq, next})
            

for comp in conns:
    recur_search(comp, {comp})
    
print(','.join(max(lanParties, key=len)))
    
```
