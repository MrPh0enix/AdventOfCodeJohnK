```python
with open('input.txt', 'r') as file:
    Doc = file.read()
    
towels = Doc.split('\n\n')[0]
towels = set(towels.split(', '))

patterns = Doc.split('\n\n')[1].split('\n')
maxlen = max([len(x) for x in towels])
print(len(patterns))
print(maxlen)
```

```python
from functools import cache

@cache
def checkPossible(pattern):
    idx = 0
    if pattern == '':
        return True
    while idx <= min(len(pattern), maxlen):
        if pattern[:idx] in towels and checkPossible(pattern[idx:]):
            return True
        idx += 1
    return False

numPats = 0

for pattern in patterns:
    
    if checkPossible(pattern):
        numPats += 1
        
print(f'Ans: {numPats}')
    
        
```

```python
from functools import cache

@cache
def checkPossible(pattern):
    idx = 0
    count = 0
    if pattern == '':
        return 1
    while idx <= min(len(pattern), maxlen):
        if pattern[:idx] in towels:
            count += checkPossible(pattern[idx:])
        idx += 1
    return count

numPats = 0

for pattern in patterns:
    numPats += checkPossible(pattern)
        
print(f'Ans: {numPats}')
```
