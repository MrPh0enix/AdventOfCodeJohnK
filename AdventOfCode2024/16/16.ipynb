```python
with open('input2.txt', 'r') as file:
    Doc = file.read()
    
Lis = Doc.split('\n')
finLis = [list(i) for i in Lis]
for idx1, i in enumerate(finLis):
    for idx2, j in enumerate(i):
        if j == 'S':
            print(f'START: ({idx1}, {idx2})')
            start = (idx1, idx2)
        elif j == 'E':
            print(f'FINISH: ({idx1}, {idx2})')
            finish = (idx1, idx2)
```

```python
import heapq

def shortestPathDijkstras(mapLis, start = start, end = finish, dir = (0, 1)):
    
    queue = [(0, start, dir)]
    
    distDict = {}
    for y in range(len(mapLis)):
        for x in range(len(mapLis[0])):
            distDict[(y, x)] = 1e13
    distDict[start] = 0
    
    visited = []
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    while queue:
        
        curr_dist, (curr_y, curr_x), curr_dir = heapq.heappop(queue)
        
        if (curr_y, curr_x) == end:
            print(distDict[(1, 13)])
            return curr_dist
        
        if (curr_y, curr_x) in visited:
            continue
        
        visited.append((curr_y, curr_x))
        
        for dy, dx in directions:
            
            next_y, next_x = curr_y + dy, curr_x + dx
            
            if mapLis[next_y][next_x] != '#':
                
                if curr_dir == (dy, dx):
                    new_dist = curr_dist + 1
                elif (dy, dx) == tuple(i * -1 for i in curr_dir):
                    new_dist = curr_dist + 2000 + 1
                else:
                    new_dist = curr_dist + 1000 + 1
                
                
                if  distDict[(next_y, next_x)] > new_dist:
                    distDict[(next_y, next_x)] = new_dist
                    heapq.heappush(queue, (new_dist, (next_y, next_x), (dy, dx))) 
    
    return 1e9


print(shortestPathDijkstras(mapLis = finLis))
    
    
```

```python
import heapq
shortestPathNodes = set()

def shortestPathDijkstras_part2(mapLis, start = start, end = finish, dir = (0, 1)):
    
    global shortestPathNodes
    
    queue = [(0, start, dir)]
    
    distDict = {}
    for y in range(len(mapLis)):
        for x in range(len(mapLis[0])):
            distDict[(y, x)] = (1e13, [])
    distDict[start] = (0, [])
    
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    while queue:
        
        curr_dist, (curr_y, curr_x), curr_dir = heapq.heappop(queue)
        
        for dy, dx in directions:
            
            next_y, next_x = curr_y + dy, curr_x + dx
            
            if mapLis[next_y][next_x] != '#':
                
                if curr_dir == (dy, dx):
                    new_dist = curr_dist + 1
                elif (dy, dx) == tuple(i * -1 for i in curr_dir):
                    # new_dist = curr_dist + 2000 + 1
                    continue
                else:
                    new_dist = curr_dist + 1000 + 1
                
                    
                if distDict[(next_y, next_x)][0] > new_dist:

                    distDict[(next_y, next_x)] = (new_dist, [(curr_y, curr_x)])
                    heapq.heappush(queue, (new_dist, (next_y, next_x), (dy, dx)))
                elif distDict[(next_y, next_x)][0] == new_dist:
                        
                    distDict[(next_y, next_x)][1].append((curr_y, curr_x))
                    heapq.heappush(queue, (new_dist, (next_y, next_x), (dy, dx)))
    
                   
    def recur_add(distDict, node, start):
        
        if node == start:
            return
        
        for i in distDict[node][1]: 
            shortestPathNodes.add(i)  
            recur_add(distDict, i, start)
    
    recur_add(distDict, finish, start)
    shortestPathNodes.add(start)
    shortestPathNodes.add(end)
    
    return len(shortestPathNodes)


print(shortestPathDijkstras_part2(mapLis = finLis))
    
```
