```python
with open('input.txt', 'r') as file:
    Doc = file.read()

Lis = Doc.split('\n\n')
initial = Lis[0]
initial = initial.split('\n')
computations = Lis[1]
computations = computations.split('\n')

vals = {}

for i in initial:
    key = i.split(': ')[0]
    val = int(i.split(': ')[1])
    vals[key] = val

# for i in computations:
#     key = i.split(' -> ')[1]
#     if key not in vals.keys():
#         vals[key] = None


    
```

```python
operations = computations.copy()


while operations != []:
    operationsDone = []
    for operation in operations:
        outKey = operation.split(' -> ')[1]
        op = operation.split(' -> ')[0]
        
        op = op.split(' ')
       
        if op[0] in vals.keys() and op[2] in vals.keys():
            
            if op[1] == 'XOR':
                vals[outKey] = vals[op[0]] ^ vals[op[2]]
                
            if op[1] == 'AND':
                vals[outKey] = vals[op[0]] & vals[op[2]]
                
            if op[1] == 'OR':
                vals[outKey] = vals[op[0]] | vals[op[2]]
                
            operationsDone.append(operation)
    
    
    for op in operationsDone:
        operations.remove(op)
    

count = 0
out = ''
while True:
    
    if count < 10:
        key = f'z0{count}'
    else:
        key = f'z{count}'
    
    if key not in vals.keys():
        break
        
    out += str(vals[key])
    count += 1
        
print(int(out[::-1], 2))

        
        
                
            
```
