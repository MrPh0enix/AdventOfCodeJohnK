```python
with open('input.txt', 'r') as file:
    Doc = file.read()

Lis = Doc.split('\n')
inptLis = [int(x) for x in Lis]
```

```python
def calc_Next(num):
    
    temp = num*64
    mixVal = temp ^ num
    pruneVal = mixVal % 16777216
    
    temp = int(pruneVal / 32)
    mixVal = temp ^ pruneVal
    pruneVal = mixVal % 16777216
    
    temp = pruneVal * 2048
    mixVal = temp ^ pruneVal
    pruneVal = mixVal % 16777216
    
    out = pruneVal
    
    return out    

secnum = 2024
for i in range(2000):
    secnum = calc_Next(secnum)
print(secnum)
```

```python

highest_diff_seq = {}

for secnum in inptLis:
    prices = [secnum % 10]
    for i in range(2000):
        secnum = calc_Next(secnum)
        prices.append(secnum % 10)
    
    seenDiff = set()
    for i in range(len(prices) - 4):
        a, b, c, d, e = prices[i], prices[i+1], prices[i+2], prices[i+3], prices[i+4]
        currDiff = (b-a, c-b, d-c, e-d)
        if currDiff in seenDiff:
            continue
        seenDiff.add(currDiff)
        if currDiff in highest_diff_seq.keys(): 
            highest_diff_seq[currDiff] += e
        else:
            highest_diff_seq[currDiff] = e

print(max(highest_diff_seq.values()))
    
```
