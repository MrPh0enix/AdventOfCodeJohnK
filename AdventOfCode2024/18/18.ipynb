```python
with open('input.txt', 'r') as file:
    Doc = file.read()
    
Lis = Doc.split('\n')
Lis = [tuple([int(_) for _ in i.split(',')]) for i in Lis]
print(len(Lis))
```

```python
import heapq

def getMap(inpt: list[tuple[int]], dims: tuple = (70, 70), steps: int = 1024):
    map = [['.' for _ in range(dims[0]+1)] for _ in range(dims[1]+1)]
    
    for i in range(steps+1):
        
        x = inpt[i][0]
        y = inpt[i][1]
        
        map[y][x] = '#'
    
    return map


start = (0, 0)
end = (70, 70)


def pathBFS(start, end, inpt: list[tuple[int]], dims: tuple = (70, 70), steps: int = 1024):
    
    map = getMap(inpt, dims, steps)
    
    visited  =[]
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    queue = [(0, start)]
    
    while queue:
        
        currLength, currPos = heapq.heappop(queue)
        
        for dir in directions:
            
            newPos = (currPos[0] + dir[0], currPos[1] + dir[1])
            
            if newPos == end:
                return currLength + 1
            
            elif newPos[0] >= 0 and newPos[0] <= dims[0] and newPos[1] >= 0 and newPos[1] <= dims[1]:
                
                if map[newPos[1]][newPos[0]] == '.' and newPos not in visited:
                    
                    visited.append(newPos)
                    heapq.heappush(queue, (currLength + 1, newPos))
    
    return -1


for idx in range(3009, 3451):
    print(idx)
    if pathBFS(start, end, Lis, steps=idx) == -1:
        print(Lis[idx])
        break
        
```
