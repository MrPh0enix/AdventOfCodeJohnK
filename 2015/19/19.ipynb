```python
with open('input.txt', 'r') as file:
    inptDoc = file.read()
    
inptLis = inptDoc.split('\n')
repDic = {}
for i in inptLis:
    if i.split(' => ')[0] not in repDic.keys():
        repDic[i.split(' => ')[0]] = [i.split(' => ')[1]]
    else:
        repDic[i.split(' => ')[0]].append(i.split(' => ')[1])

repDic
```

```python
input = 'CRnCaSiRnBSiRnFArTiBPTiTiBFArPBCaSiThSiRnTiBPBPMgArCaSiRnTiMgArCaSiThCaSiRnFArRnSiRnFArTiTiBFArCaCaSiRnSiThCaCaSiRnMgArFYSiRnFYCaFArSiThCaSiThPBPTiMgArCaPRnSiAlArPBCaCaSiRnFYSiThCaRnFArArCaCaSiRnPBSiRnFArMgYCaCaCaCaSiThCaCaSiAlArCaCaSiRnPBSiAlArBCaCaCaCaSiThCaPBSiThPBPBCaSiRnFYFArSiThCaSiRnFArBCaCaSiRnFYFArSiThCaPBSiThCaSiRnPMgArRnFArPTiBCaPRnFArCaCaCaCaSiRnCaCaSiRnFYFArFArBCaSiThFArThSiThSiRnTiRnPMgArFArCaSiThCaPBCaSiRnBFArCaCaPRnCaCaPMgArSiRnFYFArCaSiThRnPBPMgAr'
```

```python
outSet = set()

def addComb(key, start, end, inpt):
    global outSet
    if key in repDic.keys():
        lis = repDic[key]
        for i in lis:
            addStr = inpt[:start]+i+inpt[end:]
            outSet.add(addStr)


for idx, letter in enumerate(input):
    if letter.isupper():
        if input[idx+1].isupper():
            addComb(letter, idx, idx, input)
        elif input[idx+1].islower():
            addComb(input[idx:idx+2], idx, idx+2, input)

len(outSet)
```
