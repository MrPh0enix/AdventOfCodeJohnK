```python
with open('input_map.txt', 'r') as file:
    map_doc = file.read()
    
with open('input_movt.txt', 'r') as file:
    movt_doc = file.read()
    
map_lis = map_doc.split('\n')
map_lis = [list(x) for x in map_lis]

movt_lis = list(movt_doc)

for idx1, i in enumerate(map_lis):
    for idx2, j in enumerate(i):
        if j == '@':
            print(f'(y, x): ({idx1}, {idx2})')
```

```python
import copy



def moveSeq(map_lis, movt_lis):
    
    map_lis_ = copy.deepcopy(map_lis)
    
    rob_pos = (24, 24)
    
    for movt in movt_lis:
        
        if movt == '<':
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y][x-1] == 'O':
                    boxes_to_move.append((y, x-1))
                    x -= 1
                elif map_lis_[y][x-1] == '#':
                    break
                elif map_lis_[y][x-1] == '.':
                    if boxes_to_move == []:
                        map_lis_[y][x-1] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y, x-1)
                        break
                    else:
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x] = '@'
                        map_lis_[y][x+1] = '.'
                        rob_pos = (y, x)
                        for i in boxes_to_move:
                            map_lis_[i[0]][i[1]-1] = 'O'
                        break
                
            
        elif movt == '^':
            
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y-1][x] == 'O':
                    boxes_to_move.append((y-1, x))
                    y -= 1
                elif map_lis_[y-1][x] == '#':
                    break
                elif map_lis_[y-1][x] == '.':
                    if boxes_to_move == []:
                        map_lis_[y-1][x] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y-1, x)
                        break
                    else:
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x] = '@'
                        map_lis_[y+1][x] = '.'
                        rob_pos = (y, x)
                        for i in boxes_to_move:
                            map_lis_[i[0]-1][i[1]] = 'O'
                        break
            
        elif movt == 'v':
            
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y+1][x] == 'O':
                    boxes_to_move.append((y+1, x))
                    y += 1
                elif map_lis_[y+1][x] == '#':
                    break
                elif map_lis_[y+1][x] == '.':
                    if boxes_to_move == []:
                        map_lis_[y+1][x] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y+1, x)
                        break
                    else:
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x] = '@'
                        map_lis_[y-1][x] = '.'
                        rob_pos = (y, x)
                        for i in boxes_to_move:
                            map_lis_[i[0]+1][i[1]] = 'O'
                        break
            
        elif movt == '>':
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y][x+1] == 'O':
                    boxes_to_move.append((y, x+1))
                    x += 1
                    continue
                elif map_lis_[y][x+1] == '#':
                    break
                elif map_lis_[y][x+1] == '.':
                    if boxes_to_move == []:
                        map_lis_[y][x+1] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y, x+1)
                        break
                    else:
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x] = '@'
                        map_lis_[y][x-1] = '.'
                        rob_pos = (y, x)
                        for i in boxes_to_move:
                            map_lis_[i[0]][i[1]+1] = 'O'
                        break
                            
    return map_lis_

for i in moveSeq(map_lis, movt_lis):  
    print(''.join(i))

def calcGPSSum(input):
    score = 0
    for idx1, i in enumerate(input):
        for idx2, j in enumerate(i):
            if j == 'O':
                score += (idx1*100) + idx2
                
    print(score)
    

calcGPSSum(moveSeq(map_lis, movt_lis))
```

```python
def changeInputPart2(input):
    new_lis = []
    for i in input:
        row = []
        for j in i:
            if j == '#':
                row.append('#')
                row.append('#')
            elif j == 'O':
                row.append('[')
                row.append(']')
            elif j == '.':
                row.append('.')
                row.append('.')
            elif j == '@':
                row.append('@')
                row.append('.')
        new_lis.append(row)
        
    return new_lis


new_map_lis = changeInputPart2(map_lis)

for i in new_map_lis:  
    print(''.join(i))
    

for idx1, i in enumerate(new_map_lis):
    for idx2, j in enumerate(i):
        if j == '@':
            print(f'(y, x): ({idx1}, {idx2})')
```

```python
import copy

    
boxes_to_move = []  

def moveSeq_part2(new_map_lis, movt_lis):
    
    global boxes_to_move
    
    map_lis_ = copy.deepcopy(new_map_lis)
    
    rob_pos = (24, 48)
    
    for idx, movt in enumerate(movt_lis):
        print(idx)
        
        
        if movt == '<':
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y][x-1] == ']' or map_lis_[y][x-1] == '[':
                    boxes_to_move.append((y, x-1))
                    x -= 1
                elif map_lis_[y][x-1] == '#':
                    break
                elif map_lis_[y][x-1] == '.':
                    if boxes_to_move == []:
                        map_lis_[y][x-1] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y, x-1)
                        break
                    else:
                        for i in boxes_to_move[::-1]:
                            map_lis_[i[0]][i[1]-1], map_lis_[i[0]][i[1]] = map_lis_[i[0]][i[1]], map_lis_[i[0]][i[1]-1]
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x], map_lis_[y][x+1] = map_lis_[y][x+1], map_lis_[y][x]
                        rob_pos = (y, x)
                        break
                
        
        elif movt == '>':
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            while True:
                if map_lis_[y][x+1] == ']' or map_lis_[y][x+1] == '[':
                    boxes_to_move.append((y, x+1))
                    x += 1
                elif map_lis_[y][x+1] == '#':
                    break
                elif map_lis_[y][x+1] == '.':
                    if boxes_to_move == []:
                        map_lis_[y][x+1] = '@'
                        map_lis_[y][x] = '.'
                        rob_pos = (y, x+1)
                        break
                    else:
                        for i in boxes_to_move[::-1]:
                            map_lis_[i[0]][i[1]+1], map_lis_[i[0]][i[1]] = map_lis_[i[0]][i[1]], map_lis_[i[0]][i[1]+1]
                        y = boxes_to_move[0][0]
                        x = boxes_to_move[0][1]
                        map_lis_[y][x], map_lis_[y][x-1] = map_lis_[y][x-1], map_lis_[y][x]
                        rob_pos = (y, x)
                        break
                    
            
        elif movt == '^':
            
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            
            if map_lis_[y-1][x] == '.':
                map_lis_[y-1][x] = '@'
                map_lis_[y][x] = '.'
                rob_pos = (y-1, x) 
            
            elif map_lis_[y-1][x] == '#':
                pass
            
            else:
            
                def recurCheckBoxes(map_lis_, y, x):
                        global boxes_to_move
                        
                        if map_lis_[y-1][x] == '#':
                            boxes_to_move = []
                            return
                        
                        elif map_lis_[y-1][x] == '.':
                            return 
                        
                        elif map_lis_[y-1][x] == ']':
                            boxes_to_move.append((y-1, x))
                            boxes_to_move.append((y-1, x-1))
                            recurCheckBoxes(map_lis_, y-1, x)
                            recurCheckBoxes(map_lis_, y-1, x-1)
                        
                        elif map_lis_[y-1][x] == '[':
                            boxes_to_move.append((y-1, x))
                            boxes_to_move.append((y-1, x+1))
                            recurCheckBoxes(map_lis_, y-1, x)
                            recurCheckBoxes(map_lis_, y-1, x+1)
                            
                    
                recurCheckBoxes(map_lis_, y, x)
                
                if boxes_to_move == []:
                    pass
                else:
                    for i in boxes_to_move[::-1]:
                        map_lis_[i[0]-1][i[1]], map_lis_[i[0]][i[1]] = map_lis_[i[0]][i[1]], map_lis_[i[0]-1][i[1]]
                    y = boxes_to_move[0][0]
                    x = boxes_to_move[0][1]
                    map_lis_[y][x], map_lis_[y+1][x] = map_lis_[y+1][x], map_lis_[y][x]
                    rob_pos = (y, x)
                
        
            
        elif movt == 'v':
            
            boxes_to_move = []
            y = rob_pos[0]
            x = rob_pos[1]
            
            if map_lis_[y+1][x] == '.':
                map_lis_[y+1][x] = '@'
                map_lis_[y][x] = '.'
                rob_pos = (y+1, x) 
            
            elif map_lis_[y+1][x] == '#':
                pass
            
            else:
            
                def recurCheckBoxes(map_lis_, y, x):
                    
                        global boxes_to_move
                        
                        if map_lis_[y+1][x] == '#':
                            boxes_to_move = []
                            return
                        
                        elif map_lis_[y+1][x] == '.':
                            return 
                        
                        elif map_lis_[y+1][x] == ']':
                            boxes_to_move.append((y+1, x))
                            boxes_to_move.append((y+1, x-1))
                            recurCheckBoxes(map_lis_, y+1, x)
                            recurCheckBoxes(map_lis_, y+1, x-1)
                        
                        elif map_lis_[y+1][x] == '[':
                            boxes_to_move.append((y+1, x))
                            boxes_to_move.append((y+1, x+1))
                            recurCheckBoxes(map_lis_, y+1, x)
                            recurCheckBoxes(map_lis_, y+1, x+1)
                            
                    
                recurCheckBoxes(map_lis_, y, x)
                
                if boxes_to_move == []:
                    pass
                else:
                    for i in boxes_to_move[::-1]:
                        map_lis_[i[0]+1][i[1]], map_lis_[i[0]][i[1]] = map_lis_[i[0]][i[1]], map_lis_[i[0]+1][i[1]]
                    y = boxes_to_move[0][0]
                    x = boxes_to_move[0][1]
                    map_lis_[y][x], map_lis_[y-1][x] = map_lis_[y-1][x], map_lis_[y][x]
                    rob_pos = (y, x)
            
                                    
    return map_lis_


moved_map_lis_ = moveSeq_part2(new_map_lis, movt_lis) 

for i in moved_map_lis_:  
    print(''.join(i))

def calcGPSSum(input):
    score = 0
    for idx1, i in enumerate(input):
        for idx2, j in enumerate(i):
            if j in '[]':
                score += (idx1*100) + idx2
                
    print(score)
    

calcGPSSum(moved_map_lis_)
```
