```python
with open('input.txt', 'r') as file:
    Doc = file.read()
    
Lis = Doc.split('\n')
map = [list(i) for i in Lis]
for idx1, y in enumerate(map):
    for idx2, x in enumerate(y):
        if x == 'S':
            print(f'START: ({idx1}, {idx2})')
            start = (idx1, idx2)
        elif x == 'E':
            print(f'FINISH: ({idx1}, {idx2})')
            finish = (idx1, idx2)
            
```

```python
directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
path = []
visited = []
currPos = start


#to find the path
while True:
    
    if currPos == finish:
        path.append(currPos)
        break
    
    for dir in directions:
        
        new_y = currPos[0] + dir[0]
        new_x = currPos[1] + dir[1]
        
        if 0 <= new_y < len(map) and 0 <= new_x < len(map[0]) and (new_y, new_x) not in path and map[new_y][new_x] in ['.', 'E']:
            path.append((currPos[0], currPos[1]))
            
            currPos = (new_y, new_x)
            break

        
```

```python
l = 0
r = len(path) - 1
directions = [(0, 2), (2, 0), (0, -2), (-2, 0)]
count = 0


while l < len(path) - 100:
    
    while r - l >= 102:
        
        for dir in directions:
            if path[l][0] + dir[0] == path[r][0] and path[l][1] + dir[1] == path[r][1]:
                
                count += 1
        
        r -= 1
    
    l += 1
    r = len(path) - 1

print(count)
```
