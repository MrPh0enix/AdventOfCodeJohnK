```python
program = list((2,4,1,5,7,5,1,6,0,3,4,6,5,5,3,0))
A = 136904920099226
B = 0
C = 0

def exec_program(A, B, C, program):
    
    output = ''
    
    def comboOp(A, B, C, operand):
        if operand in [0, 1, 2, 3]:
            return operand
        elif operand == 4:
            return A
        elif operand == 5:
            return B
        elif operand == 6:
            return C
    
    idx = 0
    
    while idx < len(program)-1:
        opcode = program[idx]
        operand = program[idx+1]
        
        if opcode == 0:
            out = A
            den = comboOp(A, B, C, operand)
            out >>= den
            A = int(out)
            
        elif opcode == 1:
            out = B ^ operand
            B = out
        
        elif opcode == 2:
            out = comboOp(A, B, C, operand) % 8
            B = out
            
        elif opcode == 3:
            if A == 0:
                pass
            else:
                idx = operand
                continue
        
        elif opcode == 4:
            out = B ^ C
            B = out
        
        elif opcode == 5:
            out = comboOp(A, B, C, operand) % 8
            output += f'{out},'
            
        elif opcode == 6:
            out = A
            den = comboOp(A, B, C, operand)
            out >>= den
            B = int(out)
        
        elif opcode == 7:
            out = A
            den = comboOp(A, B, C, operand)
            out >>= den
            C = int(out) 
    
        idx += 2
    
    return output, A, B, C


output, A, B, C = exec_program(A, B, C, program)

print(f'Output: {output[:-1]}')
print(f'A: {A}')
print(f'B: {B}')
print(f'C: {C}')

```
