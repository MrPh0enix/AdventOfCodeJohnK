```python
import re

with open('input.txt', 'r') as file:
    Doc = file.read()
    
Doc = Doc.split('\n')

posVelLis = []

for i in Doc:
    pos = re.search(r"p=(.*?) v=", i).group(1)
    vel = re.search(r"v=(.*)", i).group(1)
    
    posVelLis.append([[int(pos.split(',')[0]), int(pos.split(',')[1])], (int(vel.split(',')[0]), int(vel.split(',')[1]))])
    
```

```python
import copy
import keyboard
from PIL import Image
import numpy as np


posVelLis_ = copy.deepcopy(posVelLis)

count = 0


for _ in range(9000):
    
    for idx, i in enumerate(posVelLis_):
        
        P_x = i[0][0]
        P_y = i[0][1]
        
        V_x = i[1][0]
        V_y = i[1][1]
        
        if P_x+V_x >= 101:
            P_x = (P_x+V_x) - 101
        elif P_x+V_x < 0:
            P_x = 101 + (P_x+V_x)
        else:
            P_x = P_x + V_x
            
        if P_y+V_y >= 103:
            P_y = (P_y+V_y) - 103
        elif P_y+V_y < 0:
            P_y = 103 + (P_y+V_y)
        else:
            P_y = P_y + V_y
            
        posVelLis_[idx][0][0] = P_x
        posVelLis_[idx][0][1] = P_y
        
    
    posGrid = [[0] * 101 for y in range(103)]
        
    for i in posVelLis_:
        P_x = i[0][0]
        P_y = i[0][1]
        
        posGrid[P_y][P_x] = 1
    
    count += 1
    # print(f'Corrent sec: {count}')
    
    grid_array = np.array(posGrid, dtype=np.uint8)*255
    
    image = Image.fromarray(grid_array)
    
    filename = f"Sec_{count}.png"  
    image.save(f'imgs/{filename}')

    
    # keyboard.wait('enter')
    

        
```
