```python
from itertools import permutations, product

numKeypad = {'7': (0,0), '8': (0, 1), '9': (0, 2), '4': (1, 0), '5': (1, 1), '6': (1, 2), '1': (2, 0), '2': (2, 1), '3': (2, 2), '0': (3, 1), 'A': (3, 2)}
robotAKeypad = {'^': (0, 1), 'A': (0, 2), '<': (1, 0), 'v': (1, 1), '>': (1, 2)}
robotBKeypad = {'^': (0, 1), 'A': (0, 2), '<': (1, 0), 'v': (1, 1), '>': (1, 2)}
userKeypad = {'^': (0, 1), 'A': (0, 2), '<': (1, 0), 'v': (1, 1), '>': (1, 2)}

numKStart = (3, 2)
dirKStart = (0, 2)

inputs = [ '129A', '176A', '985A', '170A', '528A' ]

inputs2 = [ '029A', '980A', '179A', '456A', '379A']
inputs3 = ['456A']
```

```python
def generateArrangements(input):
    return list(set([''.join(p) for p in permutations(input)]))

def checkDeadzone(input, initialPos, deadzone):
    out = []
    # deadzone = (0, 0)
    for i in input:
        y, x = initialPos
        flag = False
        for j in i:
            if j == '^':
               y, x = y - 1, x
               if (y, x) == deadzone:
                   flag = True
                   break
            elif j == 'v':
               y, x = y + 1, x
               if (y, x) == deadzone:
                   flag = True
                   break
            elif j == '>':
               y, x = y, x + 1
               if (y, x) == deadzone:
                   flag = True
                   break
            elif j == '<':
               y, x = y, x - 1
               if (y, x) == deadzone:
                   flag = True
                   break
               
        if not flag:
            out.append(i+'A')
            
    return out
            

def convertRobotAKeypad(input):
    
    outs = []

    y, x = numKStart
    
    for i in input:
        end_y, end_x = numKeypad[i]
        disty, distx = end_y - y, end_x - x
        
        if disty >= 0 and distx >= 0: # end in down right
            strs = generateArrangements('v'*abs(disty) + '>'*abs(distx))
            
            strs = checkDeadzone(strs, (y, x), (3, 0))
            
            outs.append(strs)   
            y, x = end_y, end_x
            
        
        elif disty >= 0 and distx <= 0: #end in down left 
            strs = generateArrangements('<'*abs(distx) + 'v'*abs(disty))
            
            strs = checkDeadzone(strs, (y, x), (3, 0))
            
            outs.append(strs) 
            y, x = end_y, end_x
            
        
        elif disty <= 0 and distx >= 0: # end in up right
            strs = generateArrangements('>'*abs(distx) + '^'*abs(disty))
            
            strs = checkDeadzone(strs, (y, x), (3, 0))
            
            outs.append(strs)
            y, x = end_y, end_x
            
        
        elif disty <= 0 and distx <= 0: # end in up left
            strs = generateArrangements('^'*abs(disty) + '<'*abs(distx))
            
            strs = checkDeadzone(strs, (y, x), (3, 0))
            
            outs.append(strs)
            y, x = end_y, end_x
            
            
    combinations = list(product(*outs))
    combinations = [''.join(combo) for combo in combinations]
        
        
    return combinations

convertRobotAKeypad('456A')
```

```python
def convertRobotBandUserKeypad(input):
    
    outs = []
    
    y, x = dirKStart
    
    for i in input:
        end_y, end_x = robotAKeypad[i]
        disty, distx = end_y - y, end_x - x
        
        if disty >= 0 and distx >= 0: #end in down right
            strs = generateArrangements('>'*abs(distx) + 'v'*abs(disty))
            strs = checkDeadzone(strs, (y, x), (0, 0))
            outs.append(strs)   
            y, x = end_y, end_x
        
        elif disty >= 0 and distx <= 0: #end in down left
            strs = generateArrangements('v'*abs(disty) + '<'*abs(distx))
            strs = checkDeadzone(strs, (y, x), (0, 0))
            outs.append(strs)  
            y, x = end_y, end_x
            
        elif disty <= 0 and distx >= 0: # end in up right
            strs = generateArrangements('>'*abs(distx) + '^'*abs(disty))
            strs = checkDeadzone(strs, (y, x), (0, 0))
            outs.append(strs)  
            y, x = end_y, end_x
            
        elif disty <= 0 and distx <= 0: # end in up left
            strs = generateArrangements('^'*abs(disty) + '<'*abs(distx))
            strs = checkDeadzone(strs, (y, x), (0, 0))
            outs.append(strs)  
            y, x = end_y, end_x
            
    combinations = list(product(*outs))
    combinations = [''.join(combo) for combo in combinations]

    return combinations
```
