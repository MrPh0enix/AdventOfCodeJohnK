```python
numpad =  [[ '7', '8', '9'],
           ['4', '5', '6'],
           ['1', '2', '3'],
           [None, '0', 'A']]

keypad = [[None, '^', 'A'],
          ['<', 'v', '>']]

numpad_dict = {'7': (0,0), '8': (0, 1), '9': (0, 2), '4': (1, 0), '5': (1, 1), '6': (1, 2), '1': (2, 0), '2': (2, 1), '3': (2, 2), '0': (3, 1), 'A': (3, 2)}
keypad_dict = {'^': (0, 1), 'A': (0, 2), '<': (1, 0), 'v': (1, 1), '>': (1, 2)}

numKStart = (3, 2)
dirKStart = (0, 2)

inputs = [ '129A', '176A', '985A', '170A', '528A' ]
```

```python
from collections import deque
from itertools import product

def find_paths(input, keys_dict, pad):
    paths = {}
    for i in keys_dict:
        for j in keys_dict:
            if i == j: 
                paths[(i, j)] = ['A']
            else:
                q = deque([(keys_dict[i], '')])
                selectedSeqs = []
                smallestSeq = 100000
                while q:
                    DoneFlag = False
                    (y,x), currSeq = q.popleft()
                    for new_y, new_x, dirAdd in [(y-1, x, '^'), (y+1, x, 'v'), (y, x-1, '<'), (y, x+1, '>')]:
                        if new_y < 0 or new_x < 0 or new_y >= len(pad) or new_x >= len(pad[0]):
                            continue
                        elif pad[new_y][new_x] == None:
                            continue
                        elif (new_y, new_x) == keys_dict[j]:
                            if len(currSeq+dirAdd) > smallestSeq:
                                DoneFlag = True
                                break
                            selectedSeqs.append(currSeq+dirAdd+'A')
                            smallestSeq = len(currSeq+dirAdd)
                        else:
                            q.append(((new_y, new_x), currSeq+dirAdd))
                    if DoneFlag == True:
                        break
                
                paths[(i, j)] = selectedSeqs
                
                
    inputMovts = [paths[(i, j)] for i, j in zip('A'+input, input)]
    combinedInputMovts = [''.join(i) for i in product(*inputMovts)] 
    
    return combinedInputMovts
                            
paths = find_paths('029A', numpad_dict, numpad)
paths
```

```python
complexity = 0
for input in inputs:
    
    robot1 = find_paths(input, numpad_dict, numpad)
    nxt = robot1
    
    for i in range(25):
        possible_next = []
        for seq in nxt:
            possible_next += find_paths(seq, keypad_dict, keypad)
        minlen = min(map(len, possible_next))
        nxt = [x for x in possible_next if len(x) == minlen]
    
    length = len(nxt[0])
    complexity += length * int(input[:-1])
    
print(complexity)
        

        

```
