```python
from functools import lru_cache
with open('input.txt', 'r') as file:
    inputDoc = file.read()
inputLis = inputDoc.split('\n')
```

```python
inputLis2 = []

for i in inputLis:
    lis = i.split(' ')[1].split(',')
    lis = [int(j) for j in lis]
    inputLis2.append((i.split(' ')[0], lis))
```

```python
@lru_cache(maxsize = 1048576000)
def calcComb(k, v):
    
    lis = [len(a) for a in k.split('.') if a != '']
    if ",".join(str(lis)) == v:
        return True
    else:
        return False

@lru_cache(maxsize = 1048576000)
def recur(k, ctH, ctD, idx, v):
    global combCount
    
    if idx < len(k):
        
        if k[idx] == "?":

            if ctD > 0:
                k_out = k[:idx] + '.' + k[idx+1:]
                recur(k_out, ctH, ctD-1, idx+1, v)

            if ctH > 0:
                k_out = k[:idx] + '#' + k[idx+1:]
                recur(k_out, ctH-1, ctD, idx+1, v)
        
        else:
            recur(k, ctH, ctD, idx+1, v)
    
    else:
        
        if calcComb(k, v):
            
            combCount += 1
```

```python
combCount = 0

for (k, v) in inputLis2:

    ctH = sum(v) - k.count('#')
    ctD = k.count('?') - ctH
    idx = 0
    
    recur(k, ctH, ctD, idx, ",".join(str(v)))

    
print(combCount)
```

```python
#Part2
inputLis3 = []
for (k, v) in inputLis2:
    inputLis3.append((k*5, v*5))
```

```python
combCount = 0

for (k, v) in inputLis3:

    ctH = sum(v) - k.count('#')
    ctD = k.count('?') - ctH
    idx = 0
    
    recur(k, ctH, ctD, idx, ",".join(str(v)))

    
print(combCount)
```
